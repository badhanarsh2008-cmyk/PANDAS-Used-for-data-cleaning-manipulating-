# Pandas Complete Notes

Pandas is a Python library used for data analysis, data cleaning, manipulation, exploration, and preparation. It gives two main data structures:

- `Series`: one-dimensional labeled data
- `DataFrame`: two-dimensional table with rows and columns

This notebook covers the most useful pandas methods and functions with simple examples.

The standard import style is:

````python
import pandas as pd
import numpy as np
````

## 2. Series

A `Series` is like a single column. It has values and an index.

````python
marks = pd.Series([85, 90, 78, 92], index=['Aman', 'Riya', 'Kabir', 'Neha'])
marks
````

### Common Series Methods

- `head()` / `tail()`: first or last values
- `mean()`, `median()`, `sum()`, `min()`, `max()`: statistics
- `sort_values()`: sort by values
- `sort_index()`: sort by index
- `value_counts()`: count unique values
- `unique()`: unique values
- `nunique()`: number of unique values
- `isna()`: check missing values
- `fillna()`: fill missing values
- `astype()`: change data type

````python
print('Mean:', marks.mean())
print('Highest:', marks.max())
print('Sorted:')
print(marks.sort_values(ascending=False))
````

## 3. DataFrame

A `DataFrame` is a table. It contains rows and columns, similar to Excel or SQL tables.

````python
data = {
    'Name': ['Aman', 'Riya', 'Kabir', 'Neha', 'Arjun', 'Sara'],
    'Age': [21, 22, 20, 23, np.nan, 22],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Pune', 'Mumbai', None],
    'Marks': [85, 90, 78, 92, 88, np.nan],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT']
}

df = pd.DataFrame(data)
df
````

## 4. Inspecting Data

Before cleaning or analyzing data, inspect it properly.

- `head(n)`: first n rows
- `tail(n)`: last n rows
- `shape`: number of rows and columns
- `columns`: column names
- `index`: row index
- `dtypes`: data types
- `info()`: summary of columns, nulls, and data types
- `describe()`: statistics for numeric columns
- `describe(include='object')`: summary for text/categorical columns
- `sample(n)`: random rows

````python
print(df.head())
print('\nShape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nData types:')
print(df.dtypes)
print('\nStatistics:')
print(df.describe())
````

## 5. Selecting Rows and Columns

Pandas gives many ways to select data.

- `df['Column']`: select one column
- `df[['Col1', 'Col2']]`: select multiple columns
- `loc`: select by label
- `iloc`: select by position
- Boolean filtering: select rows that match a condition

````python
print(df['Name'])
print('\nMultiple columns:')
print(df[['Name', 'Marks']])
print('\nFirst two rows using iloc:')
print(df.iloc[0:2])
print('\nName and City using loc:')
print(df.loc[:, ['Name', 'City']])
print('\nStudents with marks greater than 85:')
print(df[df['Marks'] > 85])
````

## 6. Filtering Data

Use conditions to filter rows.

- `&`: AND condition
- `|`: OR condition
- `~`: NOT condition
- `isin()`: match values from a list
- `between()`: range filter
- `str.contains()`: text contains pattern

````python
print(df[(df['Department'] == 'IT') & (df['Marks'] >= 80)])
print('\nCity is Delhi or Pune:')
print(df[df['City'].isin(['Delhi', 'Pune'])])
print('\nAge between 21 and 23:')
print(df[df['Age'].between(21, 23)])
````

## 7. Data Cleaning

Data cleaning means fixing missing, duplicate, incorrect, or inconsistent data.

Important methods:

- `isna()` / `isnull()`: check missing values
- `notna()` / `notnull()`: check non-missing values
- `dropna()`: remove rows or columns with missing values
- `fillna()`: fill missing values
- `duplicated()`: check duplicate rows
- `drop_duplicates()`: remove duplicate rows
- `replace()`: replace values
- `rename()`: rename columns
- `astype()`: convert data types
- `clip()`: limit values to lower and upper bounds

````python
print('Missing values per column:')
print(df.isna().sum())

clean_df = df.copy()
clean_df['Age'] = clean_df['Age'].fillna(clean_df['Age'].median())
clean_df['Marks'] = clean_df['Marks'].fillna(clean_df['Marks'].mean())
clean_df['City'] = clean_df['City'].fillna('Unknown')

clean_df
````

## 8. String Methods

Use `.str` to apply string operations on text columns.

- `str.lower()` / `str.upper()`: change case
- `str.strip()`: remove extra spaces
- `str.contains()`: check if text contains something
- `str.replace()`: replace text
- `str.split()`: split text
- `str.len()`: text length
- `str.startswith()` / `str.endswith()`: prefix or suffix check

````python
clean_df['Name_Upper'] = clean_df['Name'].str.upper()
clean_df['City_Lower'] = clean_df['City'].str.lower()
clean_df['Name_Length'] = clean_df['Name'].str.len()
clean_df
````

## 9. Adding, Updating, and Deleting Columns

- Add column: `df['New'] = values`
- Update column: assign new values to existing column
- Delete column: `drop(columns=['Column'])`
- Insert column at position: `insert()`
- Conditional column: `np.where()` or `.loc[]`

````python
work = clean_df.copy()
work['Passed'] = work['Marks'] >= 80
work['Grade'] = np.where(work['Marks'] >= 90, 'A', np.where(work['Marks'] >= 80, 'B', 'C'))
work.insert(1, 'Student_ID', range(101, 101 + len(work)))
work
````

## 10. Sorting Data

- `sort_values(by='Column')`: sort by values
- `sort_values(by=['Col1', 'Col2'])`: sort by multiple columns
- `sort_index()`: sort by index
- `ascending=False`: descending order

````python
work.sort_values(by='Marks', ascending=False)
````

## 11. GroupBy and Aggregation

`groupby()` is used to split data into groups and calculate summaries.

Common aggregation methods:

- `sum()`
- `mean()`
- `median()`
- `min()`
- `max()`
- `count()`
- `size()`
- `agg()` for multiple aggregations

````python
print(work.groupby('Department')['Marks'].mean())

summary = work.groupby('Department').agg(
    Average_Marks=('Marks', 'mean'),
    Highest_Marks=('Marks', 'max'),
    Students=('Name', 'count')
)
summary
````

## 12. `apply`, `map`, and `applymap`

- `map()`: transform values in a Series
- `apply()`: apply a function to Series or DataFrame
- `applymap()`: apply a function to every element of a DataFrame. In newer pandas versions, prefer `DataFrame.map()` for elementwise operations.

````python
dept_labels = {'IT': 'Technology', 'HR': 'Human Resources', 'Finance': 'Accounts'}
work['Department_Full'] = work['Department'].map(dept_labels)

work['Remark'] = work['Marks'].apply(lambda x: 'Excellent' if x >= 90 else 'Good')
work[['Name', 'Department', 'Department_Full', 'Marks', 'Remark']]
````

## 13. Combining DataFrames

Important methods:

- `concat()`: stack DataFrames vertically or horizontally
- `merge()`: SQL-style joins
- `join()`: join using index

Merge join types:

- `inner`: only matching rows
- `left`: all rows from left table
- `right`: all rows from right table
- `outer`: all rows from both tables

````python
fees = pd.DataFrame({
    'Name': ['Aman', 'Riya', 'Kabir', 'Neha'],
    'Fees_Paid': [True, True, False, True]
})

merged = pd.merge(work, fees, on='Name', how='left')
merged[['Name', 'Marks', 'Fees_Paid']]
````

## 14. Reshaping Data

- `pivot()`: reshape when each row-column pair is unique
- `pivot_table()`: pivot with aggregation
- `melt()`: convert wide data to long data
- `stack()` / `unstack()`: reshape indexes
- `transpose()` or `.T`: switch rows and columns

````python
pivot = work.pivot_table(values='Marks', index='Department', columns='Grade', aggfunc='mean')
print(pivot)

long_data = work.melt(id_vars=['Name'], value_vars=['Age', 'Marks'], var_name='Feature', value_name='Value')
long_data.head()
````

## 15. Date and Time Data

Pandas is very useful for time series data.

- `pd.to_datetime()`: convert to datetime
- `dt.year`, `dt.month`, `dt.day`: extract parts
- `dt.day_name()`: day name
- `date_range()`: create dates
- `resample()`: group time series by date frequency

````python
sales = pd.DataFrame({
    'Date': pd.date_range('2026-01-01', periods=6, freq='D'),
    'Sales': [1200, 1500, 900, 1800, 1600, 2100]
})

sales['Year'] = sales['Date'].dt.year
sales['Month'] = sales['Date'].dt.month
sales['Day_Name'] = sales['Date'].dt.day_name()
sales
````

## 17. Statistics and Analysis Methods

- `mean()`: average
- `median()`: middle value
- `mode()`: most common value
- `std()`: standard deviation
- `var()`: variance
- `corr()`: correlation
- `cov()`: covariance

## 19. More Missing Data Handling

- `ffill()`: fill using previous value
- `bfill()`: fill using next value
- `interpolate()`: estimate missing numeric values
- `dropna(axis=1)`: drop columns with missing values
- `thresh=n`: keep rows with at least n non-missing values

````python
temp = pd.DataFrame({'A': [1, np.nan, np.nan, 4], 'B': [10, 20, np.nan, 40]})
print('Forward fill:')
print(temp.ffill())
print('\nInterpolate:')
print(temp.interpolate())
````

## 20. Duplicate Data

Duplicate rows can affect analysis.

- `duplicated()`: returns True for duplicate rows
- `drop_duplicates()`: removes duplicate rows
- `subset=[...]`: check duplicates based on selected columns
- `keep='first'`, `keep='last'`, `keep=False`: decide which duplicates to keep

````python
dups = pd.DataFrame({'Name': ['Aman', 'Aman', 'Riya'], 'Marks': [85, 85, 90]})
print(dups.duplicated())
dups.drop_duplicates()
````

## 21. Outlier Handling

Outliers are unusually high or low values. Simple ways to handle them:

- `clip(lower, upper)`: cap values
- Use quantiles to find boundaries
- Use domain knowledge before removing data

````python
scores = pd.Series([40, 45, 50, 55, 60, 300])
lower = scores.quantile(0.05)
upper = scores.quantile(0.95)
scores_clipped = scores.clip(lower, upper)
scores_clipped
````

## 24. Mini Data Cleaning Project

This example shows a complete mini workflow:

1. Create dirty data
2. Clean column names
3. Remove spaces from text
4. Fill missing values
5. Remove duplicates
6. Create a new calculated column
7. Summarize the data

````python
dirty = pd.DataFrame({
    ' Student Name ': [' Aman ', 'Riya', 'Riya', 'Kabir', 'Neha'],
    'City ': ['Delhi', ' Mumbai ', ' Mumbai ', None, 'Pune'],
    'Math Marks': [80, 90, 90, np.nan, 95],
    'Science Marks': [85, 88, 88, 75, np.nan]
})

clean = dirty.copy()
clean.columns = clean.columns.str.strip().str.lower().str.replace(' ', '_')
clean['student_name'] = clean['student_name'].str.strip()
clean['city'] = clean['city'].str.strip().fillna('Unknown')
clean['math_marks'] = clean['math_marks'].fillna(clean['math_marks'].mean())
clean['science_marks'] = clean['science_marks'].fillna(clean['science_marks'].mean())
clean = clean.drop_duplicates()
clean['average_marks'] = clean[['math_marks', 'science_marks']].mean(axis=1)

clean
````

````python
clean.groupby('city').agg(
    students=('student_name', 'count'),
    avg_math=('math_marks', 'mean'),
    avg_science=('science_marks', 'mean'),
    avg_total=('average_marks', 'mean')
)
````

## 25. Best Practices

- Always inspect data using `head()`, `info()`, and `describe()`.
- Use `copy()` before changing a DataFrame if you want to keep the original safe.
- Check missing values using `isna().sum()`.
- Use meaningful column names.
- Prefer vectorized operations over loops.
- Use `groupby()` for summaries.
- Save cleaned data with `to_csv(..., index=False)`.
- Keep cleaning steps readable and repeatable.
