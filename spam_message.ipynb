```python
import numpy as np
import pandas as pd 
from pathlib import Path

```

```python
path = Path("spam_messages.csv")
```

```python
data = pd.read_csv(path)
df = pd.DataFrame(data)

cleandf = df.copy()

print("Columns :")
print(cleandf.columns)

print("First five values from dataframe :")
print(cleandf.head())

print("Checking null or missing values :")
print(cleandf.isna())

```

```python
#integer or numeric columns cleaning

print("Filling empty values :")

print(cleandf["num_caps"].fillna(cleandf["num_caps"].mean()))               #its just printing not replacing anything or changing anything
cleandf["num_caps"] = cleandf["num_caps"].fillna(cleandf["num_caps"].mean())    #its the main thing

print(cleandf["sender_age_days"].fillna(cleandf["sender_age_days"].median()))   #its just printing not saving any changes
cleandf["sender_age_days"] = cleandf["sender_age_days"].fillna(cleandf["sender_age_days"].median()) 

print("Cheking if any null value remaining ...")
print(cleandf.isnull().sum())

print("Duplicate check")
print(cleandf.duplicated().sum())
```

```python
#text column cleaning

cleandf["message"]= cleandf["message"].str.lower()  #equalize the format or string 

cleandf["message"] = cleandf["message"].fillna("_")

cleandf["len_msg"] = cleandf["message"].str.len()   #new columns with no. of characters in message

cleandf["word_count"] = cleandf["message"].str.split().apply(len)
print(cleandf[["message","word_count","len_msg"]].head())

print(cleandf["message"].isnull().sum())        #checks null rows
cleandf["message"].isnull().sum()

print(cleandf["message"].duplicated().sum())    #checks duplicates
cleandf["message"].duplicated().sum()
```
